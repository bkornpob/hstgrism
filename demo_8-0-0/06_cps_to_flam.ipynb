# Description
This Notebook shows how to put everything produced from earlier steps together, and finally extracts the spectrum.

Note that, in order to get the finaly result of the spectrum in the unit of flam, we need to compute:
$$ flam = cps / (apcorr * sensitivity * wwperpix) $$
where cps = counts per second (which will be computed by using cutout images produced by earlier steps), apcorr = aperture correction values (which will be computed by using hstgrism.grismapcorr.GrismApCorr class), sensitivity = sensitivity curve provided with [aXe](https://www.stsci.edu/scientific-community/software/axe), and wwperpix = wavelength per pixel (which can be computed using trace.csv produced by earlier step).

# Quick Walkthrough
- Step 1: compute cps
- Step 2: compute apcorr
- Step 3: get sensitivity
- Step 4: get wwperpix
- Step 5: get flam
- Step 6: save


```python
from hstgrism.grismapcorr import GrismApCorr
from scipy.interpolate import interp1d
from astropy.io import fits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Step 1: cps
We use cutout.fits, cutflat.fits, cutbkg.fits, mask.fits, and bbox.csv produced by earlier steps. The following cell demonstrates how to put these information together.

```python
gfile = ('./testfolder/testprefix_cutout.fits',1)
ffile = ('./testfolder/testprefix_cutflat.fits',1)
bfile = ('./testfolder/testprefix_cutbkg.fits',1)
mfile = ('./testfolder/testprefix_mask.fits',1)
bbox = pd.read_csv('./testfolder/testprefix_bbox.csv')
trace = pd.read_csv('./testfolder/testprefix_trace.csv')

ww = trace.ww.values
xh = trace.xh.values

cps = fits.open(gfile[0])[gfile[1]].data / fits.open(ffile[0])[ffile[1]].data
cps -= fits.open(bfile[0])[bfile[1]].data
cps *= fits.open(mfile[0])[mfile[1]].data
cps = np.nansum(cps,axis=0)[bbox.padxleft[0]:-bbox.padxright[0]]
plt.plot(ww,cps)
plt.xlabel('wavelength (A)',fontsize=12)
plt.ylabel('cps',fontsize=12)
plt.title('GD153',fontsize=12)
```

# Step 2: Compute apcorr
We use hstgrism.grismapcorr.GrismApCorr class to compute the aperture correction values. The following cell demonstrates how to set up, and compute the values.


```python
instrument = 'HST-WFC3-IR-G102'

tbox = pd.read_csv('./testfolder/testprefix_tbox.csv')
apsizepix = tbox.halfdyup.values + tbox.halfdylow.values + 1
apsizearcsec = apsizepix * GrismApCorr().table[instrument]['scale']

apcorrobj = GrismApCorr(instrument=instrument,apsize=apsizearcsec,wave=ww,aptype='diameter',apunit='arcsec',waveunit='A')
apcorrobj.compute()
apcorr = apcorrobj.data['apcorr']

plt.plot(ww,apcorr)
```

```python
# For comparison with our result in the previous cell, this cell shows the aperture correction values registered in GrismApCorr().table.

tmpx = GrismApCorr().table[instrument]['wave']
tmpy = GrismApCorr().table[instrument]['value']
tmpz = GrismApCorr().table[instrument]['apsize']
plt.figure(figsize=(10,10))
for i in np.arange(tmpy.shape[0]):
    plt.plot(tmpx,tmpy[i,:],label='{0} arcsec'.format(tmpz[i]))
plt.legend()
plt.xlabel('wavelength (A)',fontsize=12)
plt.ylabel('apcorr',fontsize=12)
plt.title(instrument,fontsize=12)
```

# Step 3: Get sensitivity
Sensitivity is recorded in the aXe sensitivity file. See [aXe](https://www.stsci.edu/scientific-community/software/axe) for more details. In this demo, we are extracting the first order beam of G102. The following cell demonstrates how to generate the values.

```python
sfile = ('/Users/kbhirombhakdi/_work/_calib_files/WFC3.IR.G102.cal.V4.32/WFC3.IR.G102.1st.sens.2.fits',1)
sdata = pd.DataFrame(fits.open(sfile[0])[sfile[1]].data)
smodel = interp1d(sdata.WAVELENGTH,sdata.SENSITIVITY,kind='linear',bounds_error=False,fill_value=None)
sensitivity = smodel(ww)
sensitivity[~np.isfinite(sensitivity)] = 1. # replace nan with 1. to avoid divided-by-zero problem

plt.plot(ww,sensitivity,'x:',label='from computation')
plt.errorbar(sdata.WAVELENGTH,sdata.SENSITIVITY,sdata.ERROR,alpha=0.4,label='from file')
plt.legend()
plt.xlabel('wavelength (A)',fontsize=12)
plt.ylabel('sensitivity (cps/A per flam)',fontsize=12)
plt.title(instrument,fontsize=12)
```

# Step 4: Get wwperpixel
Wavelength per pixel can be computed straightforwardly using np.diff.

```python
xhdiff = np.diff(xh)
xhdiff = np.append(xhdiff,xhdiff[-1]) # add one last element to maintain the array dimension
wwdiff = np.diff(ww)
wwdiff = np.append(wwdiff,wwdiff[-1]) # add one last element to maintain the array dimension
wwperpix = wwdiff/xhdiff

plt.plot(wwperpix)
```

# Step 5: flam = cps / (apcorr * sensitivity * wwperpix)
Simply put everything together. The following plot compares between our extraction, and GD153's spectrum (which is our object here) from the standard library [calspec](https://www.stsci.edu/hst/instrumentation/reference-data-for-calibration-and-tools/astronomical-catalogs/calspec).

```python
flam = cps / (apcorr * sensitivity * wwperpix)
```

```python
calspec = ('/Users/kbhirombhakdi/_work/_calib_files/PYSYN_CDBS/calspec/gd153_mod_010.fits',1)
caldata = pd.DataFrame(fits.open(calspec[0])[calspec[1]].data)

plt.figure(figsize=(10,10))
plt.plot(ww,flam,label='from extraction')
m = np.argwhere((caldata.WAVELENGTH.values >= ww.min())&(caldata.WAVELENGTH.values <= ww.max())).flatten()
plt.plot(caldata.WAVELENGTH.values[m],caldata.FLUX.values[m],'--',label='from calspec')
plt.legend()
plt.xlabel('wavelength (A)',fontsize=12)
plt.ylabel('flam',fontsize=12)
plt.title('GD153',fontsize=12)
plt.ylim(1e-15,5e-15)
plt.yscale('log')
```

# Step 6: save
This depends on user's choice of how the output should be saved. The following cell demonstrates how to wrap everything in a csv format.

```python
output = {'ww':ww,'cps':cps,'apcorr':apcorr,'sensitivity':sensitivity,'wwperpix':wwperpix,'flam':flam}
string = './testfolder/flam.csv'
pd.DataFrame(output).to_csv(string)
```
